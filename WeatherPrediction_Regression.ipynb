Import Libraries

```python
import pandas as pd
```

Read Dataset

```python
df = pd.read_csv('ww-ii-data.csv')
df
```

Data Preprocessiong

```python
df.info()
```

```python
df.isnull().sum()
```

Train_Test_Split

```python
X = df[['STA', 'YR', 'MO', 'DA']]
y = df['MaxTemp']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=3243)
```

Create_Regression_Model

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
model.predict([X_test.iloc[100]])
```

```python
y_pred = model.predict(X_test)
```

Model_Evaluation

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(y_test, y_pred)
```

Try_to_Upgrade_Model_Metrics

```python
df.describe()
```

Calculate_Coefficient_Of_Detemination

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
# Identify columns with non-numeric values
non_numeric_columns = df.select_dtypes(exclude=['number']).columns

# Fill NaN values with the mean, excluding non-numeric columns
for col in df.columns:
    if col not in non_numeric_columns:
        df[col] = df[col].fillna(df[col].mean())
df.isnull().sum()
```

```python
df = df.dropna(axis=1)
df.isnull().sum()
```

Again_Define_Train_Test_Split

```python
X = df[['STA', 'WindGustSpd', 'YR', 'MO', 'DA', 'SPD', 'MAX', 'MIN', 'MEA', 'SND', 'PGT', 'MeanTemp', 'MinTemp']]
y = df['MaxTemp']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=3243)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
mean_absolute_error(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)
```
